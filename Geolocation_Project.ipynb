```python
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize

client = MongoClient ('localhost', 27017)
db = client['companies']
```

Mi objetivo es ver si es posible la ubicación de nuestra empresa de videojuegos buscando el mercado europeo y localizada en algún lugar de Europa.

```python
a = db.companies.find({'$and':
                       [{'deadpooled_year':None},
                        {'offices.latitude': {'$ne': 'null'}},
                        {'offices.longitude': {'$ne': 'null'}},
                        {'number_of_employees': {'$gt': 1}},
                         {'$or':
                         [{'category_code': 'games_video'},{'category_code':'hardware'},{'category_code': 'software'},{'category_code': 'web'}]},
                        {'$or':
                         [{'offices.country_code':'ESP'},{'offices.country_code':'IRL'},{'offices.country_code':'FRA'},{'offices.country_code':'GBR'},{'offices.country_code':'GER'},{'offices.country_code':'NLD'},{'offices.country_code':'ITA'},{'offices.country_code':'PRT'},{'offices.country_code':'CHE'},{'offices.country_code':'AUT'},{'offices.country_code':'NOR'},{'offices.country_code':'SWE'}]}
                       ]}, 
                      {'name':1 , 'category_code':1, 'number_of_employees':1, 'founded_year':1, 'offices':1}).sort('number_of_employees', -1)

for _ in a:
    print("La compañía {} que se dedica a {}, tiene {} empleados.".format(_['name'], _['category_code'], _['number_of_employees']))
```

Convierto el resultado de la query en un Dataframe para poder trabajar con él y hago un normalize para que las columnas con diccionarios dentro se filtren.

```python
def database():
    cursor = db.companies.find({'$and':
                       [{'deadpooled_year':None},
                        {'offices.latitude': {'$ne': 'null'}},
                        {'offices.longitude': {'$ne': 'null'}},
                        {'number_of_employees': {'$gt': 1}},
                         {'$or':
                         [{'category_code': 'games_video'},{'category_code':'hardware'},{'category_code': 'software'},{'category_code': 'web'}]},
                        {'$or':
                         [{'offices.country_code':'ESP'},{'offices.country_code':'IRL'},{'offices.country_code':'FRA'},{'offices.country_code':'GBR'},{'offices.country_code':'GER'},{'offices.country_code':'NLD'},{'offices.country_code':'ITA'},{'offices.country_code':'PRT'},{'offices.country_code':'CHE'},{'offices.country_code':'AUT'},{'offices.country_code':'NOR'},{'offices.country_code':'SWE'}]}
                       ]}, 
                      {'name':1 , 'category_code':1, 'number_of_employees':1, 'founded_year':1, 'offices':1}).sort('number_of_employees', -1)
    df = json_normalize(data = cursor, record_path = 'offices', meta = ['name', 'category_code', 'number_of_employees', 'founded_year'])
    return df
```

```python
df = database()
df.head(20)
```

Escogemos, renombramos y eliminamos los nulos de las columnas que nos interesan. A su vez en la columna *countries* filtramos para obtener únicamente países de Europa.

```python
def cleaning(df):
    df = df[['name', 'category_code', 'founded_year', 'number_of_employees', 'country_code', 'city', 'address1', 'latitude', 'longitude']]
    df.rename(columns={'category_code': 'category', 'number_of_employees': 'number_employees', 'country_code': 'country', 'address1': 'address'}, inplace=True)
    df = df.dropna(subset=['founded_year','latitude', 'longitude'])
    countries = ['USA', 'AUS', 'CAN', 'ARG', 'BRA', 'SGP', 'JPN', 'HKG', 'SVN', 'COL', 'URY']
    df = df[~df['country'].isin(countries)].reset_index(drop=True)
    return df

df = cleaning(df)
df.head(20)
```

Hacemos una nueva columna con respecto a la variable *founded_year* para establecer si una compañía es una Startup o una Big Company.

```python
df['founded_year'].value_counts()
```

```python
def calculate_type(year):
    if year['founded_year'] >= 2008 and year['number_employees'] < 100:
        return 'Startup'
    else: return 'Big Company'
    
def get_class(df):
    df['type_company']= df.apply(calculate_type, axis=1)
    return df
df = get_class(df)
df.head(20)
```

Añadimos otra columna con la geolocalización de las oficinas.

```python
def get_geo(x,y):
    return {'type': 'Point', 'coordinates': [x,y]}
df['geopoint'] = df.apply(lambda x: get_geo(x['longitude'], x['latitude']), axis=1)
df.head()
```

Guardamos el dataframe en un json para poder importarlo a MongoDB.

```python
df.to_json('dataframe_companies.json', orient='records', lines= True)
```
